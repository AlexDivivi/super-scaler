```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import imageio
import os
from skimage.transform import rescale
from matplotlib import pyplot as plt
```

```python
class generator(nn.Module):
    def __init__(self):
        super(generator, self).__init__()
               
    def forward(self, x):
        x = self.generator(x)
        x = F.interpolate(x, scale_factor=2)
        x = F.relu(x)
        x = self.generator_2(x)
        return x
```

```python
model = torch.load('../data/models/1_1_x2_31.51714700146426.pt')
model.eval().cuda()
device =  torch.device('cuda:0')
dtype = torch.float
```

```python
def scale_image(image, scale):
    image = rescale(image=image, scale=scale, multichannel=True, anti_aliasing=True) 
    return image
```

```python
directory = f"../data/apologize/original"
lst = os.listdir(directory)
lst.sort()
lst[5300:5700]
```

```python
def load_data(start, end):
    x_images = []
    for filename in lst[start:end]:
        if filename.endswith(".jpg"):
            image = imageio.imread(f'{directory}/{filename}')
            x_images.append(scale_image(image, 1))
    data = torch.tensor(x_images, device=device, dtype=dtype)
    del x_images
    return data
```

```python
start, end, im = 0, 100, 0
for y in range(int(len(lst) / 100)):
    data = load_data(start, end)
    for i in range(data.shape[0]):
        x2 = model.forward(data[i].reshape(1, 228, 406, 3).transpose(1, 3))
        x4 = model.forward(x2)
        new = x4.transpose(1, 3).cpu().detach().numpy().reshape(912, 1624, 3)
        imageio.imwrite(f'../data/apologize/x4/{im}.jpg', new)
        im += 1
    start += 100
    end += 100
    del data
```
