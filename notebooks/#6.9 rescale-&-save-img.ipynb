```python
import imageio
import os
import operator
from skimage.transform import rescale
```

```python
arr = [0, 1, 2, 3, 4, 5, 6]
image_target_size = (150, 150)
scale = 2 # 300 / target
```

```python
def rescale_image_folder(directory, image_target_size, scale, norm_reshape = False):
    for filename in os.listdir(directory):
        if filename.endswith(".jpg"):
            image = imageio.imread(f'{directory}/{filename}')
            if norm_reshape:
                image = rescale(image=image, scale=image_target_size[0]/(image_target_size[0]*scale), multichannel=True, anti_aliasing=True)
            elif find_min_px_s(image) >= image_target_size[0] * scale:
                image = scale_down_img(image, image_target_size)
                image = cropND(image, image_target_size)
            imageio.imwrite(f'{directory}/x{scale}/{filename}', image)

def find_min_px_s(image, minima = 1000):
    x = image.shape[0]
    y = image.shape[1]
    if x <= y:
        xy = x
    else:
        xy = y
    if minima > xy:
        minima = xy
    return minima
                                
def cropND(img, image_target_size):
    start = tuple(map(lambda a, da: a//2-da//2, img.shape, image_target_size))
    end = tuple(map(operator.add, start, image_target_size))
    slices = tuple(map(slice, start, end))
    return img[slices]

def scale_down_img(img, image_target_size):
        if img.shape[0] <= img.shape[1]:
            xy = img.shape[0]
        else:
            xy = img.shape[1]
        if xy > image_target_size[0]:
            img = rescale(image=img, scale=image_target_size[0]/xy, multichannel=True, anti_aliasing=True)
        return img
```

```python
for lf in arr:
    directory = f"../data/CAT_0{lf}/x1"
    rescale_image_folder(directory, image_target_size, scale, norm_reshape = True)
```
